```python
生成器执行完毕时抛出抛出异常
生成器被销毁时抛出异常



```

```python
# 当调用next方法时, 会激活生成器, 直至遇到生成器方法的yield语句, 返回值1, 生成器方法被挂起. 此时主程序结束, 系统会自动产生一个GeneratorExit异常
def test():
    try:
        yield 1
    except GeneratorExit:
        print("error")


gen = test()
print(next(gen))
```

```python
# GeneratorExit异常只有在生成器对象被激活后，才有可能产生。至少调用一次生成器对象的next方法后，系统才会产生GeneratorExit异常
# 当一个生成器对象被销毁时，会抛出一个GeneratorExit异常。
# 显式地捕获了GeneratorExit异常。如果该异常没有被显式捕获，生成器对象也不会把该异常向主程序抛出
# 在生成器对象被销毁之前产生 GeneratorExit 异常
def test():
    try:
        yield 1
        yield 2
    except GeneratorExit:
        print('aa')


gen = test()
del gen
print('bb')
```

```python
# del gen这句，系统接到了销毁生成器的命令，在销毁之前，先扔给他一个GeneratorExit异常，这个异常被生成器捕获到，先打印aa
# 然后遇到了yield，这是生成器抛出RuntimeError异常(生成器正在消亡的过程中当然就不能迭代了)，只不过主程序ignore了这个异常了，主程序不受影响接着执行
def test():
    try:
        yield 1
        yield 2
    except GeneratorExit:
        print("aa")
        yield


gen = test()
print(next(gen))
del gen
print('bb')
```
