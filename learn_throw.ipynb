```python
"""
throw
1. 生成器的方法
2. 抛给生成器一个异常, 然后如果生成器能处理掉异常的话, throw方法接着迭代一次取得返回值
3. 捕获异常是在上一次迭代中断的位置捕获, 因为每次生成器运行的时候, 都是从上一次发生yield中断的地方开始运行


close
1. 生成器的方法
2. 向生成器抛出GeneratorExit异常, 生成器对象生命周期结束, 生成器方法后续执行的语句中, 不能再有yield语句, 否则会产生RuntimeError。
"""


def test():
    value = 1
    while True:
        try:
            print('run -- ')
            yield value
            value += 1
        except:
            value = 1


gen = test()
print(gen.send(None))
# 当抛异常时没有执行yield里的任何代码, try直接捕获的异常
print(gen.throw(Exception, "Method throw called!"))
# 抛出异常生成器并没有结束, 还可以继续执行
print(gen.send(None))
gen.close()
print('close')
print(gen.send(None))
```
